# Basics- Manipulating row, columns, indices, etc. using pandas

We start with importing some data.

```python
import pandas as pd
```

```python
df = pd.read_excel("data/Mineral data.xlsx")

df
```

# <font color = brown> A complementary look into filters </font>

## Simple filter

```python
fil = (df['mineral'] == 'ol') | (df['mineral'] == 'sp')
df[fil]
```

## Get the complementary data from the filter

```python
df[~fil]
```

# .isin()

```python
fil = df['mineral'].isin(['ol','sp','fs'])
df.loc[fil]
```

## Select datasets that contain a specific string in a series

```python
fil = df['mineral'].str.contains('fs')
df[fil]
```

# <font color = brown> An additional selection of selection options</font>


## Extract specific data ( extractData() )

```python
cIdf = pd.read_excel('data/chondrite element abundances.xlsx')
cIdf
```

```python
cIdf.loc[84, 'CM']
```

## Get the index or indices of a dataset(s)

```python
cIdf['Element'] == 'Th'
```

```python
indexEl = cIdf.index[cIdf['Element'] == 'Th'][0]

cIdf.loc[indexEl, 'CM']
```

```python
def extractData(catCol, elCol, catRow):
    indexEl = cIdf.index[cIdf[catCol] == elCol][0]
    res = cIdf.loc[indexEl, catRow]
    return res
```

```python
extractData('Element', 'Al', 'CO')
```
